### 표준화와 정규화 비교

```python
# 데이터 준비
data1 = iris.data.to_numpy() #iris.data -> Dataframe type : numpy 배열 type으로 변경
data2 = scale(data1)
data3 = minmax_scale(data1)
type(data1)
```

```python
#표준화와 정규화 결과 비교
print("d1 mean:", np.mean(data1, axis=0))
print("d1 std: ", np.std(data1, axis=0))
print("d2 mean:", np.mean(data2, axis=0))
print("d2 std: ", np.std(data2, axis=0))
print("d3 mean:", np.mean(data3, axis=0))
print("d3 std: ", np.std(data3, axis=0))
```

```python
plt.figure(figsize = (12, 4))
plt.subplot(1, 3, 1)
plt.title('Original')
plt.scatter(data1[:, 2], data1[:, 3])

plt.subplot(1, 3, 2)
plt.title('Standard_Scaled')
plt.scatter(data2[:, 2], data2[:, 3])

plt.subplot(1, 3, 3)
plt.title('Minmax_Scaled')
plt.scatter(data3[:, 2], data3[:, 3])

plt.show()
```

### Standard Scaler

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler() #객체화
scaler.fit(data1) #각 데이터 셋에 대한 평균 및 표준편차를 계산
```

```python
scaler.mean_ #각 데이터 셋에 대한 평균
```

```python
s_data = scaler.transform(data1) #각 데이터 셋에 대한 표준화된 값 계산  #변환
```

```python
plt.figure(figsize = (12, 4))

plt.subplot(1, 3, 1)
plt.title('Original')
plt.scatter(data1[:, 2], data1[:, 3])

plt.subplot(1, 3, 2)
plt.title('Standard_Scaled')
plt.scatter(s_data[:, 2], s_data[:, 3])

plt.show()
```

### 상관계수 r 구하기

```python
from numpy import mean
from numpy import std
from numpy.random import randn
from numpy.random import seed
```

```python
#seed random number generator
#데이터 샘플1
seed(1)

x = 20 * randn(1000) + 100 #1000개
y = x
plt.scatter(x, y)
plt.show()
```

```python
#수식으로 구하기
Qx = x - mean(x) #1000개
Qy = y - mean(y) #1000개
r = sum(Qx * Qy) / ((len(Qx)-1) * std(x)* std(y))
r
```

```python
#데이터 샘플2
seed(1)

x = 20 * randn(1000) + 100 #1000개
y = -3 * x
plt.scatter(x, y)
plt.show()
```

```python
#수식으로 구하기
Qx = x - mean(x) #1000개
Qy = y - mean(y) #1000개
r = sum(Qx * Qy) / ((len(Qx)-1) * std(x)* std(y))
r
```

```python
#seed random number generator
#데이터 샘플3
seed(1)

#prepare data
x = 20 * randn(1000) + 100 #1000개
y = x + (10 * randn(1000) + 50)

#summarize
print('x: mean=%.3f stdv=%.3f' % (mean(x), std(x))) #소수 3째자리까지
print('y: mean=%.3f stdv=%.3f' % (mean(y), std(y))) #소수 3째자리까지

#plot
plt.scatter(x, y)
plt.show()
```

```python
#수식으로 구하기
Qx = x - mean(x) #1000개
Qy = y - mean(y) #1000개
r = sum(Qx * Qy) / ((len(Qx)-1) * std(x)* std(y))
r
```
