### plt.plot(x축, y축)
- 제목 : plt.title('제목')
- x축 이름 설정 : plt.xlabel('x축이름')
- y축 이름 설정 : plt.ylabel('y축이름')
- 범례표시 : plt.legend()
- 그래프표시 : plt.show()
- x,y축의 비율을 동일하게 표시 : plt.axis('equl')

```python
import numpy as np
x = np.arange(1, 30)
y = x ** 3
plt.plot(x, y)
plt.show()
```

```python
x = np.arange(1,30)
y = x**2
y2 = x**2 + 5*x
y3 = x**2 + 10*x + 20
plt.plot(x, y, 'r--', y2, 'g-.', y3, 'b-') #선 모양, 색 지정
plt.show()
```

```python
import numpy as np
x = np.arange(1,30)
y = x**3
plt.plot(x,y,marker = '+', markersize = 5, color = 'b')
plt.grid(True)
plt.show()
```

```python
x = np.arange(1,30)
y = x**2
y2 = x**2 + 5*x
plt.plot(x,y,marker = '+', markersize = 5, color = 'b')
plt.plot(x,y2,marker = 'o', markersize = 5, color = 'r')
plt.grid(True)
plt.show()
```

### 히스토그램/커널 밀도 그래프
- 단변수 데이터의 분포
- displot()함수 사용
- 히스토그램 표시 X : hist = False
- 커널 밀도 그래프 표시 X : kde = False

```python
plt.figure(figsize = (5, 3)) #x축 폭, y축 높이
sns.distplot(df['petal_length'])
plt.show()
```

### 히트맵
- 2개의 범주형 변수를 각각 x, y축에 놓고 데이터를 매트릭스 형태로 분류
- heatmap() 함수를 사용

```python
d3 = d[['국어', '영어', '수학']]
plt.figure(figsize = (5, 3))
sns.heatmap(d3, annot = True, fmt = 'd', cmap = 'YlGnBu', linewidth = 0.5)
#annotate : 숫자 표시 유무, format : d(10진수)로 표시, color map : yellow~Green~Blue, linewidth : 선 굵기
plt.show()
```

### 범주형 데이터의 산점도

```python
plt.figure(figsize = (5, 5))
sns.stripplot(x = 'petal_length', y = 'petal_width', data = df, hue = 'species')
#hue가 groupby의 역할을 함
plt.show()
```

```python
plt.figure(figsize = (12, 5))
plt.subplot(1, 2, 1)
sns.stripplot(x = 'sepal_length', y = 'sepal_width', data = df, hue = 'species')

plt.subplot(1, 2, 2)
sns.stripplot(x = 'petal_length', y = 'petal_width', data = df, hue = 'species')

plt.show()
```

### 조인트 그래프
- 산점도 표시, x/y축에 각 변수에 대한 히스토그램 동시에 보여줌
- 두 변수의 관계와 데이터가 분산되어 있는 정도를 한눈에 파악하기 좋음
- jointplot() 함수 사용
- 회귀선 추가 : kind = 'reg'
- 육각 산점도 : kind = 'hex'
- 커널 밀도 그래프 : kind = 'kde'

```python
sns.jointplot(x = 'petal_length', y = 'petal_width', kind = 'reg', data = df)
plt.show()
```
